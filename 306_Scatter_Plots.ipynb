```python
#Scatter plots

import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
body_data = pd.read_csv("weights_and_heights.csv")

#Split data frames
male = body_data[body_data["Gender"] == "Male"]
female = body_data[body_data["Gender"] == "Female"]
```

```python
male_sample = male.sample(200, random_state = 42) #Random 200 sample, 42 means multiple people should get same randomly selected rows
```

```python
male_sample
```

```python
male_sample.shape
```

```python
plt.style.use("seaborn-poster")
plt.scatter(male_sample["Weight"], male_sample["Height"], color = "blue", s = 700, alpha = 0.5) #s= size of data points, alpha = transparency of data points
plt.title("Weight vs Height for Males")
plt.xlabel("Weight (lbs)")
plt.ylabel("Height (in)")
plt.tight_layout()
plt.show()
#Strong relationship between weight (X) and height (Y)
```

```python
#What if we want to show where median are?
#At V line and at H line 
plt.style.use("seaborn-poster")
plt.scatter(male_sample["Weight"], male_sample["Height"], color = "blue", s = 700, alpha = 0.5) #s= size of data points, alpha = transparency of data points
plt.title("Weight vs Height for Males")
plt.xlabel("Weight (lbs)")
plt.ylabel("Height (in)")

plt.axvline(x = male_sample["Weight"].median(), color = "black", linestyle = "--" ) #Median
plt.axhline(y = male_sample["Height"].median(), color = "black", linestyle = "--" ) #Median
plt.tight_layout()
plt.show()
#Strong relationship between weight (X) and height (Y)
```
